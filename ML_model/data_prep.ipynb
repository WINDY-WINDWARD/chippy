```python
# load emotion data from csv files

import pandas as pd

tes = pd.read_csv('./test.csv')
trn = pd.read_csv('./training.csv')
valid = pd.read_csv('./validation.csv')
```

```python
tes.head()
```

```python
trn.head()
```

```python
valid.head()
```

```python
# remap
# 0 - 2
# 1 - 1
# 2 - 3
# 3 - 4
# 4 - 5
# 5 - 0

trn['emotion'] = trn['label'].map({0:2, 1:1, 2:3, 3:4, 4:5 , 5:0})
tes['emotion'] = tes['label'].map({0:2, 1:1, 2:3, 3:4, 4:5, 5:0})
valid['emotion'] = valid['label'].map({0:2, 1:1, 2:3, 3:4, 4:5, 5:0})

```

```python
trn.to_csv('./training_preped.csv', index=False)
trn.head()
```

```python
tes.to_csv('./test_preped.csv', index=False)
tes.head()
```

```python
valid.to_csv('./validation_preped.csv', index=False)
valid.head()
```

```python
# drop label column and cast emotion to int
trn = trn.drop(columns=['label'])
trn['emotion'] = trn['emotion'].astype(int)
trn.head()
```

```python
tes = tes.drop(columns=['label'])
tes['emotion'] = tes['emotion'].astype(int)
tes.head()


valid = valid.drop(columns=['label'])
valid['emotion'] = valid['emotion'].astype(int)
valid.head()
```

```python
original = pd.read_csv('./emotion_data_prep.csv')
original.head()
```

```python
emotion_data = pd.concat([original, trn, tes, valid])
```
